```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

img = Image.open("cameraman.png")

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
img_array = np.array(img)

top_left = img_array[0:100, 0:100]

plt.imshow(top_left, cmap='gray')
plt.title("Top Left 100x100")
plt.axis('off')
plt.show()
```

```python
img_rgb = img.convert("RGB")
img_array = np.array(img_rgb)

R = img_array[:, :, 0]
G = img_array[:, :, 1]
B = img_array[:, :, 2]

plt.imshow(R, cmap='Reds')
plt.title("Red Channel")
plt.show()

plt.imshow(G, cmap='Greens')
plt.title("Green Channel")
plt.show()

plt.imshow(B, cmap='Blues')
plt.title("Blue Channel")
plt.show()
```

```python
modified = img_array.copy()

modified[0:100, 0:100] = 210

plt.imshow(modified)
plt.title("Modified Image")
plt.axis('off')
plt.show()
```

```python
gray_img = Image.open("lena_gray.gif")

plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
gray_array = np.array(gray_img)

h, w = gray_array.shape

center = gray_array[h//2 - 75:h//2 + 75, w//2 - 75:w//2 + 75]

plt.imshow(center, cmap='gray')
plt.title("Center 150x150")
plt.axis('off')
plt.show()
```

```python
binary = gray_array.copy()

binary[binary < 100] = 0
binary[binary >= 100] = 255

plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
plt.show()
```

```python
rotated = gray_img.rotate(-90)  # negative = clockwise

plt.imshow(rotated, cmap='gray')
plt.title("Rotated Image")
plt.axis('off')
plt.show()
```

```python
rgb_img = gray_img.convert("RGB")

plt.imshow(rgb_img)
plt.title("RGB Image")
plt.axis('off')
plt.show()
```
